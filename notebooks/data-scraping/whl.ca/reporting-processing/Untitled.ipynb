```python
"""
Team Stats Per Game
"""

SQL = """
create table reporting.whl_team_stats_by_game as
select a.season_id as season_id,
       a.id as game_id, 
       a.home_team as team_id,
       a.visiting_team as opponent_id,
       a.home_faceoffs_att as faceoffs_att,
       a.home_faceoffs_won as faceoffs_won,
       a.home_goal_count as goal_count,
       a.home_goal_total as goal_total,
       a.home_inf_count as inf_count,
       a.home_pim_total as pim_total,
       a.home_points_count as points_count,
       a.home_powerplay_count as powerplay_count,
       a.home_powerplay_goals as powerplay_goals,
       a.home_shootout as shootout,
       a.home_total_hits as total_hits,
       a.home_total_shots as total_shots,
       a.home_total_shots_on as total_shots_on
from whl_game_summary_meta a
left join whl_team_schedule_by_season b 
    on a.home_team = b.team_id
    and a.season_id = CAST(b.season_id AS INT)
    and a.id = CAST(b.game_id AS INT)
UNION
select a.season_id as season_id,
       a.id as game_id, 
       a.visiting_team as team_id,
       a.home_team as opponent_id,
       a.visiting_faceoffs_att as faceoffs_att,
       a.visiting_faceoffs_won as faceoffs_won,
       a.visiting_goal_count as goal_count,
       a.visiting_goal_total as goal_total,
       a.visiting_inf_count as inf_count,
       a.visiting_pim_total as pim_total,
       a.visiting_points_count as points_count,
       a.visiting_powerplay_count as powerplay_count,
       a.visiting_powerplay_goals as powerplay_goals,
       a.visiting_shootout as shootout,
       a.visiting_total_hits as total_hits,
       a.visiting_total_shots as total_shots,
       a.visiting_total_shots_on as total_shots_on
from whl_game_summary_meta a
left join whl_team_schedule_by_season b 
    on a.visiting_team = b.team_id
    and a.season_id = b.season_id
    and a.id = b.game_id;
"""
```
