```python
from sqlalchemy import create_engine
import pandas as pd
import numpy as np
```

```python
engine = create_engine('postgresql+psycopg2://hitmen_hockey:hitmen_hockey@localhost/Hitmen_Hockey')
```

```python
df = pd.read_sql_query('select * from hitmen_hockey',con=engine)
df_players = pd.read_sql_query('select * from hitmen_hockey_players',con=engine)
df_games = pd.read_sql_query('select * from games',con=engine)
```

```python
df_players_list = df_players[['player_id', 'player_name', 'player']].values.tolist()
df_games_list = df_games.values.tolist()
```

```python
main_list = []
for game in df_games_list:
    for player in df_players_list:
        main_list.append([game[0], player[0], player[1], player[2]])
```

```python
df_main = pd.DataFrame(main_list)
df_main.columns = ['game_id', 'player_id', 'player_name', 'player']
```

```python
# Shots Attempt For
df_shotsattemptfor = df[df['ClipType'] == 'Shot Attempt For'].groupby(['Game', 'Scouted Player']).count().sort_values('Game')
df_shotsattemptfor = df_shotsattemptfor.reset_index()

df_shotsattemptfor = df_shotsattemptfor[['Game', 'Scouted Player', 'index']]
df_shotsattemptfor.columns = ['game','scouted player', 'shotsattemptfor']

df_main = pd.merge(df_main, df_shotsattemptfor, left_on=['player', 'game_id'], right_on=['scouted player', 'game'], how='left')
df_main = df_main.drop(['game', 'scouted player'], axis=1)
```

```python
# Shots on Goal
df_shotsongoal = df[(df['ClipType'] == 'Shot Attempt For') & (df['SAGRADE'] == '2 - On Net')].groupby(['Game', 'Scouted Player']).count().sort_values('Game')
df_shotsongoal = df_shotsongoal.reset_index()

df_shotsongoal = df_shotsongoal[['Game', 'Scouted Player', 'index']]
df_shotsongoal.columns = ['game','scouted player', 'shotsongoal']

df_main = pd.merge(df_main, df_shotsongoal, left_on=['player', 'game_id'], right_on=['scouted player', 'game'], how='left')
df_main = df_main.drop(['game', 'scouted player'], axis=1)
```

```python
# Chance For
df_chance_for = df[(df['ClipType'] == 'Chance For')].groupby(['Game', 'Scouted Player']).count().sort_values('Game')
df_chance_for = df_chance_for.reset_index()

df_chance_for = df_chance_for[['Game', 'Scouted Player', 'index']]
df_chance_for.columns = ['game','scouted player', 'chancefor']

df_main = pd.merge(df_main, df_chance_for, left_on=['player', 'game_id'], right_on=['scouted player', 'game'], how='left')
df_main = df_main.drop(['game', 'scouted player'], axis=1)
```

```python
# Chance Created
df_chance_created = df[(df['CHANCE CREATE'] != 'ChanceCreate') & (df['CHANCE CREATE'].notnull())].groupby(['Game', 'CHANCE CREATE']).count().sort_values('Game')
df_chance_created = df_chance_created.reset_index()

df_chance_created = df_chance_created[['Game', 'CHANCE CREATE', 'index']]
df_chance_created.columns = ['game','scouted player', 'chancecreate']

df_main = pd.merge(df_main, df_chance_created, left_on=['player', 'game_id'], right_on=['scouted player', 'game'], how='left')
df_main = df_main.drop(['game', 'scouted player'], axis=1)
```

```python
# Turnover
df_turnover = df[df['ClipType'] == 'Turnover'].groupby(['Game', 'Scouted Player']).count().sort_values('Game')
df_turnover = df_turnover.reset_index()

df_turnover = df_turnover[['Game', 'Scouted Player', 'index']]
df_turnover.columns = ['game','scouted player', 'turnover']

df_main = pd.merge(df_main, df_turnover, left_on=['player', 'game_id'], right_on=['scouted player', 'game'], how='left')
df_main = df_main.drop(['game', 'scouted player'], axis=1)
```

```python
# Takeaway
df_takeaway = df[df['ClipType'] == 'Takeaway'].groupby(['Game', 'Scouted Player']).count().sort_values('Game')
df_takeaway = df_takeaway.reset_index()

df_takeaway = df_takeaway[['Game', 'Scouted Player', 'index']]
df_takeaway.columns = ['game','scouted player', 'takeaway']

df_main = pd.merge(df_main, df_takeaway, left_on=['player', 'game_id'], right_on=['scouted player', 'game'], how='left')
df_main = df_main.drop(['game', 'scouted player'], axis=1)
```

```python
# Zone Entry
df_zoneentry = df[df['ClipType'] == 'Zone Entry'].groupby(['Game', 'Scouted Player']).count().sort_values('Game')
df_zoneentry = df_zoneentry.reset_index()

df_zoneentry = df_zoneentry[['Game', 'Scouted Player', 'index']]
df_zoneentry.columns = ['game','scouted player', 'zoneentry']

df_main = pd.merge(df_main, df_zoneentry, left_on=['player', 'game_id'], right_on=['scouted player', 'game'], how='left')
df_main = df_main.drop(['game', 'scouted player'], axis=1)
```

```python
# Zone Exit
df_zoneexit = df[df['ClipType'] == 'Zone Exit'].groupby(['Game', 'Scouted Player']).count().sort_values('Game')
df_zoneexit = df_zoneexit.reset_index()

df_zoneexit = df_zoneexit[['Game', 'Scouted Player', 'index']]
df_zoneexit.columns = ['game','scouted player', 'zoneexit']

df_main = pd.merge(df_main, df_zoneexit, left_on=['player', 'game_id'], right_on=['scouted player', 'game'], how='left')
df_main = df_main.drop(['game', 'scouted player'], axis=1)
```

```python
# Hit
df_hit = df[df['ClipType'] == 'HIT'].groupby(['Game', 'Scouted Player']).count().sort_values('Game')
df_hit = df_hit.reset_index()

df_hit = df_hit[['Game', 'Scouted Player', 'index']]
df_hit.columns = ['game','scouted player', 'hit']

df_main = pd.merge(df_main, df_hit, left_on=['player', 'game_id'], right_on=['scouted player', 'game'], how='left')
df_main = df_main.drop(['game', 'scouted player'], axis=1)
```

```python
# Draw Penalty
df_drawpenalty = df[df['ClipType'] == 'Draw Penalty'].groupby(['Game', 'Scouted Player']).count().sort_values('Game')
df_drawpenalty = df_drawpenalty.reset_index()

df_drawpenalty = df_drawpenalty[['Game', 'Scouted Player', 'index']]
df_drawpenalty.columns = ['game','scouted player', 'drawpenalty']

df_main = pd.merge(df_main, df_drawpenalty, left_on=['player', 'game_id'], right_on=['scouted player', 'game'], how='left')
df_main = df_main.drop(['game', 'scouted player'], axis=1)
```

```python
# Faceoff Win
df_faceoffwin = df[df['FOResult'] == 'win'].groupby(['Game', 'Scouted Player']).count().sort_values('Game')
df_faceoffwin = df_faceoffwin.reset_index()

df_faceoffwin = df_faceoffwin[['Game', 'Scouted Player', 'index']]
df_faceoffwin.columns = ['game','scouted player', 'faceoffwin']

df_main = pd.merge(df_main, df_faceoffwin, left_on=['player', 'game_id'], right_on=['scouted player', 'game'], how='left')
df_main = df_main.drop(['game', 'scouted player'], axis=1)
```

```python
# Faceoff Loss
df_faceoffloss = df[df['FOResult'] == 'loss'].groupby(['Game', 'Scouted Player']).count().sort_values('Game')
df_faceoffloss = df_faceoffloss.reset_index()

df_faceoffloss = df_faceoffloss[['Game', 'Scouted Player', 'index']]
df_faceoffloss.columns = ['game','scouted player', 'faceoffloss']

df_main = pd.merge(df_main, df_faceoffloss, left_on=['player', 'game_id'], right_on=['scouted player', 'game'], how='left')
df_main = df_main.drop(['game', 'scouted player'], axis=1)
```

```python
df_main = df_main.fillna(0)
```

```python
df_main.game_id = df_main.game_id.astype(int)

df_main.shotsattemptfor = df_main.shotsattemptfor.astype(int)
df_main.shotsongoal = df_main.shotsongoal.astype(int)
df_main.chancefor = df_main.chancefor.astype(int)
df_main.chancecreate = df_main.chancecreate.astype(int)
df_main.turnover = df_main.turnover.astype(int)
df_main.takeaway = df_main.takeaway.astype(int)
df_main.zoneentry = df_main.zoneentry.astype(int)
df_main.zoneexit = df_main.zoneexit.astype(int)
df_main.hit = df_main.hit.astype(int)
df_main.blockedshot = df_main.blockedshot.astype(int)
df_main.drawpenalty = df_main.drawpenalty.astype(int)
df_main.faceoffwin = df_main.faceoffwin.astype(int)
df_main.faceoffloss = df_main.faceoffloss.astype(int)
```

```python
df_main.to_sql("game_totals_individual", con=engine, if_exists='replace')
```
